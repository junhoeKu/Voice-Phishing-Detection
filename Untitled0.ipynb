# Sliding Window

```python
window_size = 1024
overlap = 256
stride = window_size - overlap # 다음 윈도우 시작 위치까지의 거리

new_rows = []

for index, row in korccvi.iterrows():
    text = row['Transcript_Cleaned']
    label = row['Label']
    text_len = len(text)

    # 텍스트가 윈도우 크기보다 작거나 같으면 원본 그대로 추가
    if text_len <= window_size:
        new_rows.append({ 'Transcript_Chunk': text, 'Label': label})
        continue

    # 텍스트를 슬라이딩하며 분할
    start = 0
    while start < text_len:
        end = start + window_size
        chunk = text[start:end]

        # 마지막 청크가 윈도우 크기보다 작을 경우, 텍스트의 끝에서부터 window_size만큼 잘라냄
        if len(chunk) < window_size:
             last_chunk_start = max(0, text_len - window_size)
             chunk = text[last_chunk_start:]
             new_rows.append({'Transcript_Chunk': chunk, 'Label': label})
             break # 루프 종료

        new_rows.append({'Transcript_Chunk': chunk, 'Label': label})
        start += stride

df_sliding_window = pd.DataFrame(new_rows)
```

```python
df_sliding_window.to_csv('/content/drive/MyDrive/Github/2025_Voicephishing/korccvi_sliding_window.csv', index=False)
```

# Head & Tail

```python
head_size = 512
tail_size = 512
total_size = head_size + tail_size

def create_head_tail_text(text):
    """텍스트의 앞/뒤를 잘라 재구성하는 함수"""
    if len(text) > total_size:
        return text[:head_size] + text[-tail_size:]
    else:
        return text

# 원본 데이터프레임을 복사하여 새로운 컬럼을 추가
df_head_tail = korccvi.copy()
df_head_tail['text'] = df_head_tail['text'].apply(create_head_tail_text)

# 가독성을 위해 원본 Transcript_Cleaned 열은 제외하고 출력
df_head_tail = df_head_tail[['text', 'Label']]
```
