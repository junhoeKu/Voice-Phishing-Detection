```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSequenceClassification
from peft import PeftModel
import gradio as gr
import torch
import os
import torch.nn.functional as F
from sentence_transformers import SentenceTransformer
from huggingface_hub import HfApi, HfFolder

# ===== 고정 경로 설정 =====
BASE_DIR = "/content/drive/MyDrive/Github/2025_Voicephishing"
CACHE_DIR = os.path.join(BASE_DIR, "cache")
BASE_MODEL_PATH = "Qwen/Qwen2.5-0.5B-Instruct"
# (중요) 훈련된 '분류' 어댑터 경로
ADAPTER_PATH = os.path.join(BASE_DIR, "model/model_qwen_cls_bt_all_512_25")
TOKENIZER_PATH = os.path.join(BASE_MODEL_PATH)

# ===== 환경 변수 설정 (원본 유지) =====
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["TRANSFORMERS_CACHE"] = CACHE_DIR

# ... (기타 환경 변수) ...
hf_token = ""
os.environ["HUGGINGFACE_TOKEN"] = hf_token
HfFolder.save_token(hf_token)

# ===== GPU 설정 =====
device = "cuda" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32

# ===================================================================
#  1. 모델 1: '분류기' (Classifier) 불러오기
# (훈련(Training) 때와 동일한 AutoModelForSequenceClassification 사용)
# ===================================================================
print("Loading Classifier (Model 1)...")
classifier_model = AutoModelForSequenceClassification.from_pretrained(
    BASE_MODEL_PATH,
    num_labels=2, # 훈련 시 2로 설정했음
    torch_dtype=torch_dtype,
    cache_dir=CACHE_DIR,
    low_cpu_mem_usage=True
)
# PEFT 모델(어댑터) 불러오기
classifier_model = PeftModel.from_pretrained(classifier_model, ADAPTER_PATH)

# (선택적) 모델 병합. 추론 속도가 빨라짐
classifier_model = classifier_model.merge_and_unload()
classifier_model.eval()
classifier_model.to(device)
print("Classifier (Model 1) loaded.")

# ===================================================================
#  2. 모델 2: '설명기' (Generator) 불러오기
# (훈련되지 않은 *원본* CausalLM 모델)
# ===================================================================
print("Loading Generator (Model 2)...")
generator_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_PATH,
    torch_dtype=torch_dtype,
    cache_dir=CACHE_DIR,
    low_cpu_mem_usage=True
)
generator_model.eval()
generator_model.to(device)
print("Generator (Model 2) loaded.")


# ===== 토크나이저 불러오기 (공통 사용) =====
tokenizer = AutoTokenizer.from_pretrained(TOKENIZER_PATH, cache_dir=CACHE_DIR, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'left' # জenerate를 위해 padding side를 left로 설정

## 입력 문장 전처리
def preprocess_input(text):
    text = " ".join(text.strip().split())
    return text

# ===================================================================
#  3. 추론 파이프라인 (두 모델을 모두 사용)
# ===================================================================
def generate_response(prompt):

    # --- 3A. '모델 1 (분류기)'로 분류 수행 ---
    processed_prompt = preprocess_input(prompt)
    inputs = tokenizer(
        processed_prompt,
        return_tensors="pt",
        truncation=True,
        max_length=1024,
        padding=True
    ).to(device)

    with torch.no_grad():
        outputs = classifier_model(**inputs)
        logits = outputs.logits
        probabilities = F.softmax(logits, dim=1)
        prediction_idx = torch.argmax(probabilities, dim=1).item()

    if prediction_idx == 1:
        classification_result = "보이스피싱"
        confidence = probabilities[0][1].item()
    else:
        classification_result = "정상 대화"
        confidence = probabilities[0][0].item()

    # --- 3B. '모델 2 (설명기)'로 이유 생성 (⭐프롬프트 최적화⭐) ---

    # 1. 역할(Persona) - 더 강력하고 구체적으로
    role = "당신은 보이스피싱 텍스트를 분석하는 AI 수사관입니다."

    # 2. 맥락(Context) - 1번 모델의 결론을 명확히 전달
    context = f"다음 텍스트는 1차 분석 결과 '{classification_result}'로 판별되었습니다."

    # 3. 임무(Task) - '이유'를 '패턴'에 근거하여 '검증'하고 '설명'하도록 지시
    task = f"이 텍스트를 정밀 분석하여, '{classification_result}' 판별이 타당한지 검증하고, 그 핵심 근거를 '보이스피싱 패턴'(기관 사칭, 긴급성, 금전/정보 요구 등)에 기반하여 설명하세요."

    # 4. 출력 포맷(Format) - 일관된 결과를 위해
    output_format = "반드시 [판단 근거] 3개와 [결론]으로 구성된 보고서 형식으로 답변하고, 정상 대화라면 정상 대화인 이유를 분석해서 작성하세요."

    # 5. 최종 결합 (더 간결한 f-string here-doc 방식 사용)
    reasoning_prompt = f"""{role}
{context}

[임무]
{task}
{output_format}

[분석 대상 텍스트]
{processed_prompt}

[분석 보고서]
"""

    # 설명기 토크나이징
    gen_inputs = tokenizer(reasoning_prompt, return_tensors="pt").to(device)

    # 생성 (이전과 동일)
    gen_outputs = generator_model.generate(
        **gen_inputs,
        max_new_tokens=200,
        min_new_tokens=30,
        num_beams=3,
        do_sample=True,
        top_p=0.8,
        temperature=0.7,
        no_repeat_ngram_size=3,
        repetition_penalty=1.5,
        early_stopping=True,
        pad_token_id=tokenizer.eos_token_id,
    )

    # 생성된 텍스트 디코딩 (입력 프롬프트 제외)
    generated_text = tokenizer.decode(gen_outputs[0][gen_inputs.input_ids.shape[1]:], skip_special_tokens=True).strip()

    # --- 3C. 최종 결과 포맷팅 ---
    final_output = f"## 1. 분석 결과:\n"
    final_output += f"**{classification_result}** (신뢰도: {confidence:.2%})\n\n"
    final_output += f"## 2. 분석 보고서 (AI 수사관):\n"
    final_output += f"{generated_text}" # 모델이 생성한 [판단 근거]와 [결론]

    return final_output

## Gradio 인터페이스 생성 (Markdown 출력을 위해 gr.Markdown 사용)
interface = gr.Interface(
    fn=generate_response,
    inputs=[gr.Textbox(lines=10, placeholder="분석할 대화를 입력하세요...")],
    outputs=gr.Markdown(label="분석 결과"), # Textbox 대신 Markdown 사용
    title="보이스피싱 탐지 AI (분류 및 분석)",
    description="입력한 대화를 분석하여 보이스피싱 여부를 '분류'하고, '이유'를 생성합니다.",
)

## Gradio 앱 실행
if __name__ == "__main__":
    interface.launch(share=True)
```
