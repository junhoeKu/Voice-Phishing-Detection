```python
import os
import pandas as pd
from datasets import Dataset
import random

def load_voicephishing_data(file_path: str):
    if file_path.endswith(".csv"):
        df = pd.read_csv(file_path)
    else:
        raise ValueError("CSV ÌååÏùºÎßå ÏßÄÏõêÎê©ÎãàÎã§.")

    # ÏûÖÎ†• ÌÖçÏä§Ìä∏: ÎåÄÌôî ÎÇ¥Ïö©
    df["text"] = df["text"].astype(str)

    df = df[["text", "label"]].copy()
    dataset = Dataset.from_pandas(df)

    # ÌïôÏäµ, Í≤ÄÏ¶ù, ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖã Î∂ÑÌï†
    train_val_split = dataset.train_test_split(test_size=0.3, shuffle=True, seed=42)
    val_test_split = train_val_split["test"].train_test_split(test_size=0.5, shuffle=True, seed=42)

    # Í∞Å Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ ÌååÏùºÎ°ú Ï†ÄÏû•ÌïòÏó¨ Ïû¨ÌòÑÏÑ± ÌôïÎ≥¥
    save_dir = os.path.dirname(file_path)
    train_val_split['train'].to_csv(os.path.join(save_dir, 'spam_stt_train.csv'), index=False)
    val_test_split['train'].to_csv(os.path.join(save_dir, 'spam_stt_val.csv'), index=False)
    val_test_split['test'].to_csv(os.path.join(save_dir, 'spam_stt_test.csv'), index=False)

    return train_val_split['train'], val_test_split['train'], val_test_split['test']

if __name__ == "__main__":
    data_file_path = "/content/drive/MyDrive/Github/2025_Voicephishing/dataset/spam_stt.csv"
    try:
        load_voicephishing_data(data_file_path)
        print("üéâ Îç∞Ïù¥ÌÑ∞ÏÖã Î∂ÑÌï† Î∞è Ï†ÄÏû•Ïù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§. Ïù¥Ï†ú train.pyÎ•º Ïã§ÌñâÌï† Ïàò ÏûàÏäµÎãàÎã§.")
    except FileNotFoundError:
        print(f"‚ùå Ïò§Î•ò: ÏßÄÏ†ïÎêú Í≤ΩÎ°úÏóê ÌååÏùºÏù¥ ÏóÜÏäµÎãàÎã§. Í≤ΩÎ°úÎ•º ÌôïÏù∏Ìï¥ Ï£ºÏÑ∏Ïöî: {data_file_path}")
```
